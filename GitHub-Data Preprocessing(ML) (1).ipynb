# Seperating Dependent and Independent Variables

```python
X = df.iloc[:, 1:-1].values
y = df.iloc[:, -1].values
```

```python
X
```

```python
y
```

# Handling the Missing Values

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')#filling the missing values by average or mean
imputer.fit(X[:, 1:3])
X[:, 1:3] = imputer.transform(X[:, 1:3])
```

# Encoding the Categorical Data

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer(transformers=[('encoding', OneHotEncoder(),[0])],remainder = 'passthrough')
X = np.array(ct.fit_transform(X))
```

```python
X
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
y
```

# Dividing the Testset and Trainingset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:,3:5] = sc.fit_transform(X_train[:, 3:5])
X_test[:, 3:5] = sc.fit_transform(X_test[:, 3:5])
```
